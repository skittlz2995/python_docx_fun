```python
import os
import docx

directory = "/Users/naa0290/Documents/Personal/documents_final"
output_directory = "/Users/naa0290/Documents/Personal/documents_final/output"
if not os.path.exists(output_directory):
    os.mkdir(output_directory)

for filename in os.listdir(directory):
    if filename.endswith(".docx") and not filename.startswith("~"):
        docname = os.path.join(directory, filename)
        print("Parsing "+filename)
        doc = docx.Document(docname)

        key = 'Tech Support 360'
        val = 'Tech 360'

        # Create list of all paragraphs
        paragraphs = list(doc.paragraphs)

        # Add table paragraphs
        for table in doc.tables:
            for row in table.rows:
                for cell in row.cells:
                    for paragraph in cell.paragraphs:
                        paragraphs.append(paragraph)

        # Add header/footer paragraphs
        for section in doc.sections:            
            for paragraph in section.header.paragraphs:
                paragraphs.append(paragraph)
            for paragraph in section.even_page_header.paragraphs:
                paragraphs.append(paragraph)
            for paragraph in section.first_page_header.paragraphs:
                paragraphs.append(paragraph)
            for paragraph in section.footer.paragraphs:
                paragraphs.append(paragraph)
            for paragraph in section.even_page_footer.paragraphs:
                paragraphs.append(paragraph)
            for paragraph in section.first_page_footer.paragraphs:
                paragraphs.append(paragraph)
            
        for p in paragraphs:
            if key in p.text:
                inline = p.runs
                for i in range(len(inline)):
                    # Replace all single runs
                    if inline[i].text.find(key) != -1:
                        text = inline[i].text.replace(key, str(val))
                        inline[i].text = text
                    
                    # Find partial if not on last line
                    if len(inline[i].text)>0 and inline[i].text[-1] in key and i != len(inline)-1:
                        check_length = len(inline[i].text)
                        start_index = check_length - len(key)
                        for text_index in range(start_index, check_length):
                            if inline[i].text[text_index:check_length] not in key:
                                continue
                            else:
                                chars_found = check_length - text_index
                                chars_remain = len(key) - chars_found
                                if inline[i].text[text_index:check_length] != key[0:chars_found]:
                                    # Mismatch
                                    break
                                # Check for full match across two inlines
                                if inline[i+1].text[0:len(key)-chars_found] == key[chars_found:check_length]:
                                    # Remove partial key from inline i and inline i+1
                                    inline[i].text = inline[i].text[0:text_index]
                                    inline[i+1].text = inline[i+1].text[chars_remain:len(inline[i+1].text)]

                                    # Add val to beginning of second inline
                                    inline[i+1].text = val + inline[i+1].text
                                
                                break
                                
        doc.save(os.path.join(directory, "output", "parsed_"+filename))
        print('Done')
    else:
        continue
                 
            
```
